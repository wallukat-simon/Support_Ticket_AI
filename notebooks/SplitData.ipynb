### Display the original dataset
Notice that the original dataset contains entries in english as well as german. For this usecase, it is better to use a single language at the time, so the preprocessing analysis as well as the trained model later on will be more robust. 

```python
import pandas as pd

# Path to original dataset
path = "../data/dataset-tickets.csv"
# Read dadaset csv file and store it as DataFrame
df = pd.read_csv(path)
df
```

## Split the dataset
Remove all the rows in which the entry for the 'language' column is either 'de' or 'en'. That way, a DataFrames that contain only antries in either language will be created.

```python
# Drop rows where 'department' is 'IT'
df = df[df['language'] != 'en']

print("\nFiltered DataFrame (without IT department):")
df
```

## Save as CSV
Finally the newly created dataset is stored as a .csv file

```python
# Save to CSV
df.to_csv('../data/dataset_de.csv', index=False)
print("\nSaved to '../data/dataset_de.csv'")
```
